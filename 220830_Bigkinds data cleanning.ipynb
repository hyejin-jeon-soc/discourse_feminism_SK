```python
from gensim import corpora, models
import gensim #젠심 임포트 전에 안해서 꼭 하렴


"""
conserv10
liberal10
conserv21
liberal21

"""

# 1. conservative platforms yr from 90 to 10
conserv10_dic = corpora.Dictionary(conserv10['keyword'])
conserv10_corpus = [conserv10_dic.doc2bow(text) for text in conserv10['keyword']] #[]안 딕셔너리는 앞서 지정된 딕셔너리를 의미

print(conserv10_corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

```python
# 2. liberal platforms yr from 90 to 10

liberal10_dic = corpora.Dictionary(liberal10['keyword'])
liberal10_corpus = [liberal10_dic.doc2bow(text) for text in liberal10['keyword']] #[]안 딕셔너리는 앞서 지정된 딕셔너리를 의미

print(liberal10_corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

```python
# 3. conservative platforms yr from 11 to 21

conserv21_dic = corpora.Dictionary(conserv21['keyword'])
conserv21_corpus = [conserv21_dic.doc2bow(text) for text in conserv21['keyword']] #[]안 딕셔너리는 앞서 지정된 딕셔너리를 의미

print(conserv21_corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

```python
# 4. liberal platforms yr from 11 to 21

liberal21_dic = corpora.Dictionary(liberal21['keyword'])
liberal21_corpus = [liberal21_dic.doc2bow(text) for text in liberal21['keyword']] #[]안 딕셔너리는 앞서 지정된 딕셔너리를 의미

print(liberal21_corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
```

# PPL score & coherence Score to select Topic N = k

```python
'''
https://brunch.co.kr/@jiheon0105/38 
참조링크

'''


import gensim 
from gensim.models import CoherenceModel
Lda = gensim.models.ldamodel.LdaModel # 아래 for문 앞에 쓰임 c


# 1. conserv10 
con10_ppl_score = []
con10_cohere_score = []

## range from 2 to 30, 31 is not included
for i in range(2,31):
    conserv10_lda = Lda(conserv10_corpus, num_topics = i, id2word= conserv10_dic, passes = 100 , iterations = 300,  chunksize = 150, random_state = 0, alpha = 'auto')
    con10_ppl_score.append(conserv10_lda.log_perplexity(conserv10_corpus)) #혼잡도 아래 일관도
    con10_cohere_score.append(CoherenceModel(model = conserv10_lda, texts = conserv10['keyword'] , dictionary = conserv10_dic, coherence='c_v', topn = 10).get_coherence())
    print(i, 'the process complete')
    
    
```

```python
# 2. liberal10 
lib10_ppl_score = []
lib10_cohere_score = []

## range from 2 to 30, 31 is not included
for i in range(2,31):
    liberal10_lda = Lda(liberal10_corpus, num_topics = i, id2word= liberal10_dic, passes = 100 , iterations = 300,  chunksize = 150, random_state = 0, alpha = 'auto')
    lib10_ppl_score.append(liberal10_lda.log_perplexity(liberal10_corpus)) #혼잡도 아래 일관도
    lib10_cohere_score.append(CoherenceModel(model = liberal10_lda, texts = liberal10['keyword'] , dictionary = liberal10_dic, coherence='c_v', topn = 10).get_coherence())
    print(i, 'the process complete')
    
```

```python
# 3. conserv21 
con21_ppl_score = []
con21_cohere_score = []

## range from 2 to 30, 31 is not included
for i in range(2,31):
    conserv21_lda = Lda(conserv21_corpus, num_topics = i, id2word= conserv21_dic, passes = 100 , iterations = 300,  chunksize = 150, random_state = 0, alpha = 'auto')
    con21_ppl_score.append(conserv21_lda.log_perplexity(conserv21_corpus)) #혼잡도 아래 일관도
    con21_cohere_score.append(CoherenceModel(model = conserv21_lda, texts = conserv21['keyword'] , dictionary = conserv21_dic, coherence='c_v', topn = 10).get_coherence())
    print(i, 'the process complete')
```

```python
# 4. liberal21 
lib21_ppl_score = []
lib21_cohere_score = []

## range from 2 to 30, 31 is not included
for i in range(2,31):
    liberal21_lda = Lda(liberal21_corpus, num_topics = i, id2word= liberal21_dic,  passes = 100 , iterations = 300,  chunksize = 150, random_state = 0, alpha = 'auto')
    lib21_ppl_score.append(liberal21_lda.log_perplexity(liberal21_corpus)) #혼잡도 아래 일관도
    lib21_cohere_score.append(CoherenceModel(model = liberal21_lda, texts = liberal21['keyword'] , dictionary = liberal21_dic, coherence='c_v', topn = 10).get_coherence())
    print(i, 'the process complete')
    
```
